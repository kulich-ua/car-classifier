```python
car_types = 'Volkswagen Golf 1', 'Volkswagen Golf 2', 'Volkswagen Golf 3'
path = Path('data/vwGolfsDDG')
# AcuraModels/AcuraModelsTrain
```

```python
if not path.exists():
    path.mkdir()
    for o in car_types:
        dest = (path/o)
        dest.mkdir(exist_ok=True)
        results = search_images_ddg(f'{o}', max_images=150)
        download_images(dest, urls=results)
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```
